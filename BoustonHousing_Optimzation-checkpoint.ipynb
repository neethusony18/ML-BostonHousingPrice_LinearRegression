```python
#Import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the database from scikit learn databases
from sklearn.datasets import load_boston
boston_df=load_boston()
boston_df.DESCR#fucntion to describe in a scikit learn dataset
#Converting it into a pandas dataframe
df=pd.DataFrame(boston_df.data,columns=boston_df.feature_names)
#df.head()#Target column is missing 
#Add the target column name
df['MEDV']=pd.Series(boston_df.target)
df.head()
```

```python
#Data wrangling:Data cleaning
df.isnull().sum()
#output shows that there are no empty columns
```

```python
#to analyse the correlation between the  independent and dependent variables
correlation_matrix=df.corr().round(2)
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(correlation_matrix,cmap="YlGnBu",annot=True,square=True,ax=ax)
```

```python
#For optimization from the previous bouston hosuing model we built,here we can consider two more features:RM which has a positive
#correlation and PTRATIO which has a negative correlation
X=pd.DataFrame(np.c_[df['RM'],df['ZN'],df['LSTAT'],df['PTRATIO']],columns=['RM','ZN','LSTAT','PTRATIO'])
Y=df['MEDV']
```

```python
#splitting the data sets into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=5)#X_train,X_test should follow this order
print("The Y_train is:",Y_train.head())
print(X_train.head())
print(X_test.head())
print("The Y_test is :",Y_test.head())
```

```python
#Building the model
from sklearn.linear_model import LinearRegression
reg_model=LinearRegression().fit(X_train,Y_train)
```

```python
#Evaluation uisng the training  data set and calculating the accuracy through the R2score
from sklearn.metrics import r2_score
Y_train_predict=reg_model.predict(X_train)
r2_train_score=r2_score(Y_train,Y_train_predict)
print("The accuracy from the training set is:",r2_train_score)
```

```python
#Evaluation uisng the testing  data set and calculating the accuracy through the R2score
from sklearn.metrics import r2_score
Y_test_predict=reg_model.predict(X_test)
r2_test_score=r2_score(Y_test,Y_test_predict)
print("The accuracy from the training set is:",r2_test_score)
```
